## Exploratory Data Analysis

```python
bar1=df['Gender'].value_counts()
bar1
```

```python
plt.bar(bar1.index, bar1.values, color= ['red','dodgerblue'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Count')
plt.figure(figsize=(10,20))
plt.show()
```

```python
sls_gen= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
a = sns.barplot(data= sls_gen,x='Gender',y= 'Amount', hue='Gender',legend=True)
plt.title('Gender Distribution')
sns.set(rc={'figure.figsize':(10,6)})

for bars in a.containers:
    a.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
plt.grid(False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
sales_age= df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False)
print(sales_age)
```

```python


sns.barplot(data= sales_age, x='Age Group', y= 'Amount',hue='Age Group')
plt.xlabel('Age Group')
plt.title('Amount Spent By Age Group')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(6)
ax= sns.barplot(data=sales_state,y='Orders',x='State', hue='State',palette='dark:dodgerblue')
plt.title('Orders By States')
plt.xlabel('States')
plt.grid(False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(6)

ax = sns.barplot(data= sales_state, x='State', y= 'Amount', hue='State' )
plt.title('Amount Spent by States')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot( data=df ,x='Marital_Status', palette='dark:dodgerblue', hue='Gender')
plt.ylabel('Count')
plt.xlabel('Marital Status')
plt.title('Marital Status distribution acco. to Gender')

sns.set(rc={'figure.figsize':(8,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(22,5)})
ax = sns.countplot(data=df, x='Product_Category',palette='dark:purple')
plt.title('Product Category Distribution')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pc= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
a = sns.barplot(data=sales_pc, x='Product_Category',y='Amount',hue='Product_Category', palette='colorblind')
sns.set(rc={'figure.figsize':(10,5)})

for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_pc= df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(6)
ap= sns.barplot(data=sales_pc, x='Product_Category',y='Orders',hue='Product_Category',palette='dark')
sns.set(rc={'figure.figsize':(10,5)})

for bars in ap.containers:
    ap.bar_label(bars)
```
